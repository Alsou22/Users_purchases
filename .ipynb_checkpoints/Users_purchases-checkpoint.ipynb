```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
                                            ПОИСК ЛОЯЛЬНЫХ ПОКУПАТЕЛЕЙ
```

```python
df = pd.read_csv('C:\\Users\\АЛСУ\\Downloads\\Новая папка (2)/lesson_3_data_1_.csv', encoding = 'Windows-1251')
df.head()
```

```python
user_df = df[['tc', 'art_sp']] # отбираем нужные нам наблюдения
user_df
```

```python
user_df.shape
```

```python
user_df = user_df.rename(columns = {'tc': 'user_id', 'art_sp': 'brand_info'})
user_df.head()
```

```python
def split_brand(brand_name_data): #нужно вывести только названия брендов, пишем функцию для разделения
    return brand_name_data.split(' ')[-1]
split_brand('MARAVILLA 500 G Store_Brand')
```

```python
user_df['brand_name'] = user_df.brand_info.apply(split_brand)
user_df.head()
```

```python
#можно так. Используем lambda функцию, чтобы разрезать строку по пробелу и взять последний элемент
user_df.brand_info.apply(lambda x: x.split(' ')[-1] )
```

```python
users_purchases = user_df \
    .groupby('user_id', as_index = False) \
    .agg({'brand_name': 'count'}) \
    .rename(columns= {'brand_name': 'purchases'})
users_purchases.head()
```

                                            ИЩЕМ ЛОЯЛЬНЫХ ПОЛЬЗОВАТЕЛЕЙ

```python
#сколько покупок совершил каждый пользователь
users_purchases = user_df \
    .groupby('user_id', as_index = False) \
    .agg({'brand_name': 'count'}) \
    .rename(columns= {'brand_name': 'purchases'})
users_purchases.head()
```

```python
users_purchases.shape #сколько уникальных пользователей
```

```python
users_purchases.purchases.median() #смотрим медиану покупок. 50% покупателей совершало больше 2-х покупок, 50% совершали меньше 2-х покупок
```

```python
users_purchases.describe()
#25 % пользователей совершили больше 5 покупок. 50% совершили больше 2-х покупок
```

```python
#отберем только тех пользователей, которые совершили  5 и более покупок
users_purchases = user_df \
    .groupby('user_id', as_index = False) \
    .agg({'brand_name': 'count'}) \
    .rename(columns= {'brand_name': 'purchases'}) \
    .query("purchases > = 5")
users_purchases.head()
```

```python
#число уникальных брендов
users_unique_brand = user_df \
    .groupby('user_id', as_index = False) \
    .agg({'brand_name': pd.Series.nunique}) \
    .rename(columns= {'brand_name': 'unique_brand'})
users_unique_brand
```

```python
users_purchases.describe()
```

```python
#Посчитаем сколько покупок каждого бренда совершил покупатель
lovely_brand_purchases_df = user_df.groupby(['user_id', 'brand_name'], as_index = False) \
    .agg({'brand_info': 'count'}) \
    .sort_values(['user_id', 'brand_info'], ascending = [False, False]) \
    .groupby('user_id') \
    .head(1) \
    .rename(columns={'brand_name': 'lovely_brand', 'brand_info': 'lovely_brand_purchases'})
lovely_brand_purchases_df
```

```python
loyalty_df = users_purchases \
    .merge(users_unique_brand, on = 'user_id') \
    .merge(lovely_brand_purchases_df, on = 'user_id')
loyalty_df.head()
#по умолчанию how = inner
```

```python
loyal_users = loyalty_df[loyalty_df.unique_brand ==1]
loyal_users
#отбираем покупателей, у которых больше 5 покупок и любимый бренд 1
```

```python
#посчитаем какая доля от всех покупок пришлась на любимый бренд у каждого покупателя
loyalty_df['loyalty_score'] = loyalty_df.lovely_brand_purchases/loyalty_df.purchases
loyalty_df
```

```python
ax = sns.distplot(loyalty_df.loyalty_score, kde = False)
```

```python
brand_loyalty = loyalty_df \
    .groupby('lovely_brand', as_index = False) \
    .agg({'loyalty_score': 'median','user_id': 'count'})
brand_loyalty
```

```python
ax = sns.barplot(x = 'lovely_brand', y = 'loyalty_score', data = brand_loyalty)
```
